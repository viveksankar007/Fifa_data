```python
import pandas as pd
fifa_df = pd.read_csv("Fifa.csv")
fifa_df.head()
```

```python
country_player_count = fifa_df['Nationality'].value_counts()

most_players_country = country_player_count.idxmax()
most_players_count = country_player_count.max()

most_players_country, most_players_count
```

```python
import matplotlib.pyplot as plt

top_5_countries = country_player_count.head(5)
plt.figure(figsize=(10,6))
top_5_countries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with the Most Number of Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

```python
fifa_df['Wage'] = fifa_df['Wage'].str.replace('€', '').str.replace('K', '').astype(float)

# Find the player with the highest salary
highest_salary_player = fifa_df.loc[fifa_df['Wage'].idxmax()]

highest_salary_player['Name'], highest_salary_player['Wage']
```

```python
plt.figure(figsize=(10,6))
plt.hist(fifa_df['Wage'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Players\' Wages')
plt.xlabel('Wage (in €K)')
plt.ylabel('Number of Players')
plt.show()
```

```python
Fifa = pd.read_csv("Fifa.csv")
Fifa
tallest_player = Fifa.sort_values(by=['Height'],ascending=False)

tallest_player['Name'],tallest_player['Height']
```

```python
club_player_count = fifa_df['Club'].value_counts()

most_players_club = club_player_count.idxmax()
most_players_club_count = club_player_count.max()

most_players_club, most_players_club_count
```

```python
preferred_foot_count = fifa_df['Preferred Foot'].value_counts()

plt.figure(figsize=(8,5))
preferred_foot_count.plot(kind='bar', color='skyblue')
plt.title('Preferred Foot of Players')
plt.xlabel('Foot')
plt.ylabel('Number of Players')
plt.xticks(rotation=0)
plt.show()
```
